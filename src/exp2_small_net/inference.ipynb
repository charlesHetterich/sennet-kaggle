```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
from torch.utils.data import DataLoader

import sys
sys.path.append('..')
import util

log_board = util.diagnostics.LogBoard('log_dir', 6005)
log_board.launch()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
test = util.data.SenNet(
    chunk_size=16,
    guarantee_vessel=0.5,
    samples=[
        "/train/kidney_2"
    ]
)

model = util.UNet3D(
    Conv3d=util.nn.Conv3DNormed,
    block_depth=8,
    batch_norm=True,
).to(device)
model.load_state_dict(torch.load("./bin/group_norm.pth"))
model.to(device)
model.eval()


agg_dice = util.AggDice(
    test.scans[0],
    test.labels[0],
)
```

```python
epochs = 10
logger = log_board.get_logger('patcher3d')
batch_size = 32

valid_data = DataLoader(test, batch_size=batch_size, shuffle=True)

with torch.no_grad():
    for epoch in range(epochs):
        for i, (x, y, pos) in enumerate(valid_data):
            step = epoch * len(test) + i
            patches_seen = step * batch_size
            x ,y = x.float().to(device), y.float().to(device)
            x = (x - x.mean()) / x.std()
            masks = model.deep_masks(y)

            p_y = [F.sigmoid(logits.float()) > 0.5 for logits in model(x)]

            seen_dice, full_dice = agg_dice(p_y[-1].cpu(), pos)
            batch_dice = util.dice_coefficient(p_y[-1], y)

            logger.add_scalars('dice', {
                'batch': batch_dice,
                'seen': seen_dice,
                'full': full_dice
            }, step)
            if step % 10 == 0:
                print('patches seen:', patches_seen, '| batch:', batch_dice.item(), '| seen:', seen_dice.item(), '| full:', full_dice.item())

            if step % 10 == 0:
                logger.add_image('masks', util.mask_plots(x, masks, p_y), step, dataformats='HWC')
```
